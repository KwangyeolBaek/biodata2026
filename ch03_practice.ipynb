### 데이터 내의 값(예: 남 vs. 여)을 따라 나누어 분석하기

```python
# 특정 조건인 데이터만 추출
data1[data1['sex']=='M']

# 참고
# data1[data1.sex=='M']
```

```python
# 특정 조건인 데이터만 추출
data1.query('sex=="M"')
```

```python
# 특정 조건인 데이터만 추출
data1[data1['sex']=='M'].describe()
```

```python
data1.groupby('sex').describe()
```

```python
data1.groupby('sex').mean(numeric_only = True)
```

```python
# 히스토그램 겹쳐 그리기
sns.histplot(data1[data1.sex=='M'].weight, color='blue')
sns.histplot(data1[data1.sex=='F'].weight, color='red')
plt.show()
```

```python
# 상자그림
sns.boxplot(x = "sex", y  = "weight", data = data1)
#sns.boxplot(x = "sex", y  = "weight", data = data1, hue = "sex")
plt.show()
```

```python
# 바이올린그림
sns.violinplot(x = "sex", y  = "weight", data = data1)
plt.show()
```

```python
# 막대그래프
sns.barplot(x = "sex", y  = "weight", data = data1)
plt.show()
```

```python
# 막대그래프: 오차막대 (error bar) 조절
sns.barplot(x = "sex", y  = "weight", data = data1, errorbar = 'sd')
plt.show()
sns.barplot(x = "sex", y  = "weight", data = data1, errorbar = 'se')
plt.show()
```

```python
# 남성/여성별로 몸무게(weight), 허리둘레(waist), 체질량지수(BMI), 혈당(Blood_glucose)의 분포를 히스토그램으로 비교
sns.histplot(data1[data1.sex=='M']['weight'], color = 'blue')
sns.histplot(data1[data1.sex=='F']['weight'], color = 'red')
plt.show()

sns.histplot(data1[data1.sex=='M']['waist'], color = 'blue')
sns.histplot(data1[data1.sex=='F']['waist'], color = 'red')
plt.show()

sns.histplot(data1[data1.sex=='M']['BMI'], color = 'blue')
sns.histplot(data1[data1.sex=='F']['BMI'], color = 'red')
plt.show()

sns.histplot(data1[data1.sex=='M']['Blood_glucose'], color = 'blue')
sns.histplot(data1[data1.sex=='F']['Blood_glucose'], color = 'red')
plt.show()
```
